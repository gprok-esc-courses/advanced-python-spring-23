# Find real unique categories

```python
categories = {}
primary_category = {}
```

```python
cat_column = df['category_name']
```

```python
for c in cat_column:
    s = str(c)
    values = s.split('/')
    if len(values) > 0 and values[0] not in primary_category:
        primary_category[values[0]] = 1
    else:
        primary_category[values[0]] += 1
    for v in values:
        if v not in categories:
            categories[v] = 1
        else:
            categories[v] += 1
        
```

```python
# Split the category_name into primary, secondary, and category
for index, row in df.iterrows():
    values = str(row['category_name']).split('/')
    if len(values) < 3:
        df.at[index, 'primary'] = 'N/A'
        df.at[index, 'secondary'] = 'N/A'
        df.at[index, 'category'] = 'N/A'
    else:
        df.at[index, 'primary'] =  values[0]
        df.at[index, 'secondary'] = values[1]
        df.at[index, 'category'] = values[2]
```

```python
test_df = pd.read_csv('data/test.tsv', sep='\t')
test_df['primary'] = None
test_df['secondary'] = None
test_df['category'] = None
for index, row in test_df.iterrows():
    values = str(row['category_name']).split('/')
    if len(values) < 3:
        test_df.at[index, 'primary'] = 'N/A'
        test_df.at[index, 'secondary'] = 'N/A'
        test_df.at[index, 'category'] = 'N/A'
    else:
        test_df.at[index, 'primary'] =  values[0]
        test_df.at[index, 'secondary'] = values[1]
        test_df.at[index, 'category'] = values[2]
```
